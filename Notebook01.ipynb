```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import metrics
```

```python
nsei_df = pd.read_csv('^NSEI.csv',index_col="Date",parse_dates=True)
nsei_df.head()
```

```python
bosch_df = pd.read_csv('BOSCHLTD.NS.csv',index_col="Date",parse_dates=True)
bosch_df.head()
```

```python
hdfc_df = pd.read_csv('HDFCBANK.NS.csv',index_col="Date",parse_dates=True)
hdfc_df.head()
```

```python
ongc_df = pd.read_csv('ONGC.NS.csv',index_col="Date",parse_dates=True)
ongc_df.head()
```

```python
fig, axs = plt.subplots(2, 2,figsize=(16,9))

axs[0, 0].plot(nsei_df['Close'],color='blue')
plt.setp(axs[0,0],ylabel="INR",title="NIFTY 50")

axs[0, 1].plot(bosch_df['Close'],color='red')
plt.setp(axs[0,1],ylabel="INR",title="BOSCHLTD NSE")

axs[1, 0].plot(ongc_df['Close'],color='green')
plt.setp(axs[1,0],ylabel="INR",title="ONGC NSE")

axs[1, 1].plot(hdfc_df['Close'])
plt.setp(axs[1,1],ylabel="INR",title="HDFCBANK NSE")
```
